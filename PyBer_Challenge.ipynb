# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load 
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
total_rides_per_citytype = pyber_data_df.groupby(['type']).count()['ride_id']
total_rides_per_citytype
```

```python
# 2. Get the total drivers for each city type
total_drivers_per_citytype = city_data_df.groupby(['type']).sum()['driver_count']
total_drivers_per_citytype
```

```python
#  3. Get the total amount of fares for each city type
total_amount_of_fares_per_citytype = pyber_data_df.groupby(['type']).sum()['fare']
total_amount_of_fares_per_citytype
```

```python
#  4. Get the average fare per ride for each city type.
sum_fares_by_type = pyber_data_df.groupby(['type']).sum()['fare']
total_avg_of_fares_per_citytype = sum_fares_by_type / total_rides_per_citytype
total_avg_of_fares_per_citytype
```

```python
# 5. Get the average fare per driver for each city type.
sum_fares_by_type = pyber_data_df.groupby(['type']).sum()['fare']
total_avg_of_fares_per_driver= sum_fares_by_type / total_drivers_per_citytype
total_avg_of_fares_per_driver
```

```python
#  6. Create a PyBer summary DataFrame. 

pyber_summary_df = pd.DataFrame({
    "Total Rides" : total_rides_per_citytype,
    "Total Drivers" : total_drivers_per_citytype,
    "Total Fares" : total_amount_of_fares_per_citytype,
    "Average Fare per Ride" : total_avg_of_fares_per_citytype,
    "Average Fare per Driver" : total_avg_of_fares_per_driver
})
pyber_summary_df  
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
#  8. Format the columns.

pyber_summary_df["Total Rides"] = pyber_summary_df["Total Rides"].map('{:,}'.format)
pyber_summary_df["Total Drivers"] = pyber_summary_df["Total Drivers"].map('{:,}'.format)
pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map('${:,.0f}'.format)
pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map('${:,.2f}'.format)
pyber_summary_df["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].map('${:,.2f}'.format)

pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df.head()
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.

total_fare_per_city_and_date_df = pyber_data_df.groupby(['type','date']).sum()['fare']
total_fare_per_city_and_date_df.head()
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
total_fare_per_city_and_date_df = total_fare_per_city_and_date_df.reset_index()
total_fare_per_city_and_date_df.head(10)
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 

total_fare_per_city_and_date_df['date'] = pd.to_datetime(total_fare_per_city_and_date_df['date'])
total_fare_per_city_and_date_df_pivot = total_fare_per_city_and_date_df.pivot(index="date", columns= "type", values="fare")
total_fare_per_city_and_date_df_pivot.head(10)
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
total_fare_per_city_and_date_df_1Q_2019 = total_fare_per_city_and_date_df_pivot.loc['2019-01-01':'2019-04-28']
total_fare_per_city_and_date_df_1Q_2019
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)

total_fare_per_city_and_date_df_1Q_2019.index = pd.to_datetime(total_fare_per_city_and_date_df_1Q_2019.index)
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
total_fare_per_city_and_date_df_1Q_2019.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.

total_fare_per_city_and_date_W_df = total_fare_per_city_and_date_df_1Q_2019.resample('W').sum()
total_fare_per_city_and_date_W_df.head(10)
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 
import matplotlib.dates as mdates
#  Plot the resample DataFrame 
fig, ax = plt.subplots(figsize = (15,6))
ax.plot(total_fare_per_city_and_date_W_df)

# Add a title 
ax.set_title("Total Fare by City Type") 

# Add a x- and y-axis label.
#total_fare_per_city_and_date_W_df.set_xlabel("Week") 
#ax.set_xlabel("Week")
ax.set_xticks(pd.date_range(start = "2019-01-01", end = "2019-04-28", freq="MS"))
ax.set_ylabel('Fare ($USD)',fontsize=14) 
ax.legend(["Rural","Suburban","Urban"])
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b'))
# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

```
